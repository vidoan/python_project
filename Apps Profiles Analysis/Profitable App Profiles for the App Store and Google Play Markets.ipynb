```python
def is_english(string):
    '''
    The numbers corresponding to the characters we commonly use in an English text are all in the range 0 to 127
    We loop each character in the string to identify any character that is non-english.
    If the input string has more than three characters that fall outside the ASCII range (0 - 127), that string is non-english
    '''
    non_english_characters = 0
    for character in string:
        if ord(character) > 127:
            non_english_characters +=1
        if non_english_characters > 3:
            return False
    return True

print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print('\n')
print(is_english('Instachat üòú'))
print('\n')
print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
```

### Find the most common apps by genre

```python
## create a function to generate frequency tables that show percentages

def freq_table(dataset, index):
    table = {}
    total = 0

    for row in dataset:
        total+=1
        value = row[index]
        if value in table:
            table[value]+=1
        else:
            table[value]=1
            
    table_percentage = {}
    
    for key in table:
        percentage = (table[key]/total) *100
        table_percentage[key] = percentage
        
    return table_percentage
            
```

```python
## create a function to sort table by percent in descending order
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
display_table(free_english_iOS_app, 11)
```

### Find the most common apps by genre in App Store

The frequency tables we analyzed on the previous screen showed us that apps designed for fun dominate the App Store, while Google Play shows a more balanced landscape of both practical and fun apps. Now, we'd like to determine the kind of apps with the most users.

One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. For the Google Play data set, we can find this information in the Installs column, but this information is missing for the App Store data set. As a workaround, we'll take the total number of user ratings as a proxy, which we can find in the `rating_count_tot` app.

```python
#calculate the average number of user ratings for each genre
#generate frequecy table for prime_genre
prime_genre =freq_table(free_english_iOS_app, -5)

for genre in prime_genre:
    total = 0
    len_genre = 0
    for app in free_english_iOS_app:
        genre_app = app[-5]
        user_ratings = float(app[5])
        if genre_app == genre:
            total+= user_ratings
            len_genre+= 1
    average_n_rating = total/len_genre
    print(f'{genre}: {average_n_rating}')
            
```

We're going to leave the numbers as they are, which means that we'll consider that an app with 100,000+ installs has 100,000 installs, and an app with 1,000,000+ installs has 1,000,000 installs, and so on. To perform computations, however, we'll need to convert each install number from a string to a float. This means we need to remove the commas and the plus characters, or the conversion will fail and cause an error.

```python
categories_android = freq_table(free_english_android_app, 1)
best_category = ''
highest_n_installs = 0

for category in categories_android:
    total = 0
    len_category = 0
    for app in free_english_android_app:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    
    if avg_n_installs > highest_n_installs:
        best_category = category
        highest_n_installs = avg_n_installs
    print(category, ':', avg_n_installs)
print(f'Best category: {best_category}')
print(f'Number of installs: {highest_n_installs}')
```

On average, communication apps have the most installs: 38,456,119. This number is heavily skewed up by a few apps that have over one billion installs (WhatsApp, Facebook Messenger, Skype, Google Chrome, Gmail, and Hangouts), and a few others with over 100 and 500 million installs:

```python
for app in free_english_android_app:
    if app[1] == 'COMMUNICATION':
        print(f'{app[0]} : {app[5]}')
```
